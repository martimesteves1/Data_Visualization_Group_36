```python
athlete_events = pd.read_csv('athlete_events.csv')
display(athlete_events.head(5))
country_definitions = pd.read_csv('country_definitions.csv')
display(country_definitions.head(5))
gdp = pd.read_csv('gdp_csv.csv')
display(gdp.head(5))
population = pd.read_csv('population.csv')
display(population.head(5))
country_iso = pd.read_csv('country_iso.csv', header=1, index_col=0)
display(country_iso.head(5))
iso_regions = pd.read_csv('iso_regions.csv')
display(iso_regions.head(5))
```

```python
indexes = list(country_definitions[~(country_definitions['region'].isin(country_iso['Countries and Areas of Interest']))].index)
indexes
```

```python
country_definitions[~(country_definitions['region'].isin(country_iso['Countries and Areas of Interest']))]
```

```python
country_definitions_clean = country_definitions[['NOC', 'region']].copy().rename(columns={'region': 'Country'})
country_definitions_clean.iloc[indexes,1] = ['Curaçao', 'Bolivia', 'Republic of the Congo', 'Russian Federation', 'United Kingdom', ' Individual Olympic Athletes',\
                                              'United States Virgin Islands', 'Russian Federation', 'Saint Kitts and Nevis', 'Trinidad and Tobago',\
                                              'Russian Federation', 'United States of America', 'West Indies Federation']
```

```python
country_ioc_iso = country_definitions_clean.join(country_iso[['Countries and Areas of Interest', 'Alpha-3']].drop_duplicates().set_index('Countries and Areas of Interest'), on='Country')
country_ioc_iso
```

```python
country_ioc_iso_region = country_ioc_iso.join(iso_regions[['alpha-3', 'region', 'sub-region']].set_index('alpha-3'), on='Alpha-3')
country_ioc_iso_region[country_ioc_iso_region['region'].isna()]
```

```python
athlete_events_codes = athlete_events.join(country_ioc_iso.set_index('NOC'), on='NOC').rename(columns={'Alpha-3': 'ISO'})
athlete_events_codes = athlete_events_codes.join(iso_regions[['alpha-3' , 'region', 'sub-region']].set_index('alpha-3'), on='ISO').dropna(subset='sub-region')
display(athlete_events_codes)
athlete_events_codes.to_csv('athlete_events_codes.csv')
```

```python
medals = athlete_events_codes.copy()
medals = medals.drop_duplicates(subset=['Year','Team', 'Event', 'Medal'])
medals_gender = medals.groupby(['Season', 'Year', 'NOC', 'Sex']).size().unstack('Sex').fillna(0)
medals_gender['Total'] = medals_gender['F'] + medals_gender['M']
medals_gender['Difference'] = medals_gender['M'] - medals_gender['F']
medals_gender['Difference ABS'] = abs(medals_gender['M'] - medals_gender['F'])
medals_gender['Difference %'] = medals_gender['Difference']/medals_gender['Total'].replace(0, np.nan)
medals_gender['Difference % ABS'] = medals_gender['Difference ABS']/medals_gender['Total'].replace(0, np.nan)
display(medals_gender)
```

```python
medals_gender = medals_gender.reset_index().join(country_ioc_iso_region.set_index('NOC'), on='NOC')
medals_gender.to_csv('medals_gender.csv')
```

```python
df_gender_subset = df_gender.iloc[:,[1] + list(range(198, 198+32))]
df_gender_subset
```

```python
gender_data_final = pd.DataFrame()
gender_data_final['Year'] = np.tile(np.arange(1990, 2022), country_ioc_iso_region.shape[0])
gender_data_final['Country'] = np.repeat(country_ioc_iso_region['Country'], 32).reset_index(drop=True)
```

```python
df1_long = df_gender_subset.melt(id_vars='country', var_name='Year', value_name='HDI').rename(columns={'country': 'Country'})
df1_long['Year'] = df1_long['Year'].str.extract('(\d+)').astype(int)  # Extract year as integer
merged_df = pd.merge(gender_data_final, df1_long, on=['Country', 'Year'], how='left')
merged_df.to_csv('gender_data.csv')
```
